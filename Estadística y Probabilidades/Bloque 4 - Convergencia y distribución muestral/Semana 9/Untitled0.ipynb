```python
import random

# Simular 1000 lanzamientos de dos dados y sumar sus resultados
resultados = [random.randint(1,6) + random.randint(1,6) for _ in range(1000)]

# Calcular la media muestral
media_muestral = sum(resultados) / len(resultados)
print(f"Media muestral: {media_muestral}")
```
